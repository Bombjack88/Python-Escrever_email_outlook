# Desafio Python e E-mail


Digamos que trabalhas em uma indústria e estás responsável pela área de inteligência de negócio.

Todos os dias, a equipe ou até mesmo um programa, gera um report diferente para cada área da empresa:
- Financeiro
- Logística
- Manutenção
- Marketing
- Operações
- Produção
- Vendas

Cada um desses reports deve ser enviado por e-mail para o Gerente de cada Área.
A relação de Gerentes (com seus respectivos e-mails) e áreas está no arquivo 'Enviar E-mails.xlsx'.

Crie uma automação em python para envio dos ficheiros.

# Resolução 1

```python
import pandas as pd
import win32com.client as win32
```

```python
gerentes_df = pd.read_excel('Enviar E-mails.xlsx')
display(gerentes_df)
```

```python
outlook = win32.Dispatch('outlook.application')

# utilizo o enumerate para obter o index e depois com o .loc identifico a célula específica

for i, email in enumerate(gerentes_df['E-mail']):
    gerente=gerentes_df.loc[i,'Gerente']
    relatorio=  gerentes_df.loc[i,'Relatório']
                            
    mail = outlook.CreateItem(0) 
    mail.To = email
    mail.Subject = f'Relatório {relatorio}'
    
    mail.Body = f'''
    Prezado {gerente}, 
    Segue em anexo o Relatório de {relatorio}, conforme solicitado.
    Qualquer dúvida estou à disposição.
    Att.,
    '''
    # Anexos (posso colocar quantos quiser):
    
    attachment  = r'C:\Users\fiu126\Documents\MeusProjetos\Python-Escrever_email_outlook\{}.xlsx'.format(relatorio)
    mail.Attachments.Add(attachment)

    mail.Display(True) # -> apenas para ver o e-mail
    #mail.Send() ->enviar o e-mail (bloqueado)
```
